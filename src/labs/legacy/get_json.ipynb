```python
# import requests
# from urllib import request, urlopen
import requests
import urlopen
import pandas as pd
import numpy as np
```

```python
url = "https://api.exchangeratesapi.io/history?start_at=2014-10-20&end_at=2018-09-01&base=USD&symbols=IDR"

response = requests.request("GET", url)
```

```python
response.json()
```

```python
report = pd.io.json.json_normalize(response.json()['rates'])
```

```python
data =report.transpose()
```

```python
data.sort_index()
```

```python
data_array = np.array(data)
```

```python
data = pd.DataFrame(data_array)
```
