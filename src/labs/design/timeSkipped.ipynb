```python
import random
import datetime as dt
import numpy as np
import pandas as pd
```

```python
def generate_row(year, month, day):
    while True:
        date = dt.datetime(year, month, day)
        data = np.random.random(size=4)
        yield [date]+list(data)
```

```python
# Days i Have data for
dates = [
    (2000,1,1),
    (2000,1,2),
    (2000,1,4),
]
generators = [generate_row(*date) for date in dates]
```

```python
# get 5 data points for each
data = [
    next(generator) for generator in generators for _ in range(5)
]
```

```python
df = pd.DataFrame(
    data, 
    columns=['dates'] + ['f'+str(i) for i in range(1,5)],
    
)
```

```python
df
```

```python
# df group by

groupby_day = df.groupby(pd.PeriodIndex(data=df.dates, freq='D'))
results = groupby_day.sum()
```

```python
idx = pd.date_range(min(df.dates), '2000-01-06')
results.reindex(idx, fill_value=0)
```

```python
results.set_index(results.index, inplace=True)
results.resample('D').sum().fillna(19)
```
