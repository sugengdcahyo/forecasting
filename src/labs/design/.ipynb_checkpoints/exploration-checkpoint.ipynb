```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
DATASETS_BASE_DIR = '../datasets/'
USD_DIR = DATASETS_BASE_DIR+'usd'
JPY_DIR = DATASETS_BASE_DIR+'jpy'
```

```python
try:
    USD = os.path.join(DATASETS_BASE_DIR, 'USD_IDR Historical Data.csv')
    JPY = os.path.join(DATASETS_BASE_DIR, 'JPY_IDR Historical Data.csv')
except:
    print("dataset not found.")
```

```python
USD = pd.read_csv(USD, sep=',')
JPY = pd.read_csv(JPY, sep=',')
```

# <b>Current Exchange</b>

## **Data preprocessing**
### A. USD Process
#### Data cleaning and standarization
- [x] Convert object to time in date column

```python
usd = USD
usd['Date'] = usd.Date.apply(
    lambda x: pd.to_datetime(x).strftime('%Y-%m-%d')
)
usd
```

- [x] Convert object to time in date column
- [x] Remove comma value on exchange rate
- [x] Remove percent value on exchange rate

```python
# replace commas symbol
usd = usd.replace(',', '', regex=True)
# clear % symbol
usd = usd.replace('%', '', regex=True)

usd
```

### B. JPY Process
#### Data cleaning and standarization
- [x] Convert object to time in date column

```python
jpy = JPY
jpy['Date'] = jpy.Date.apply(
    lambda x: pd.to_datetime(x).strftime('%Y-%m-%d')
)
jpy
```

- [x] Convert object to time in date column
- [x] Remove comma value on exchange rate
- [x] Remove percent value on exchange rate

```python
# replace commas symbol
jpy = jpy.replace(',', '', regex=True)
# clear % symbol
jpy = jpy.replace('%', '', regex=True)

jpy
```

## **Plotting Data to CandleCharts**

```python
from plotly import graph_objects as go
```

```python
def figureCandleStick(dataframe):
    fig = go.Figure(
        data=[go.Candlestick(
            x=dataframe['Date'],
            open=dataframe['Open'],
            high=dataframe['High'],
            low=dataframe['Low'],
            close=dataframe['Price']
        )]
    )
    
    fig.update_layout(
        autosize=True,
        height=800,
        xaxis_rangeslider_visible=False
    )
    
    fig.show(config = dict({'scrollZoom': True}))
```

### A. Figure USD-IDR

```python
figureCandleStick(usd)
```

### B. Figure JPY-IDR

```python
figureCandleStick(jpy)
```
